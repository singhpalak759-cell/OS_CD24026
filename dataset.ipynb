```python
import random
import pandas as pd
import numpy as np
import string
import struct
import sys


cities = ["Nagpur", "Pune", "Mumbai", "Delhi", "Bangalore"]
weather = ["Clear", "Rainy", "Cloudy"]
traffic = ["Low", "Medium", "High"]
food_type = ["Veg", "Fast Food", "Dessert", "Healthy"]

def generate_dataset(n=100):

    dataset = []

    for i in range(n):

        record = {
            "Order_ID": i+1,
            "City": random.choice(cities),
            "Delivery_Distance_km": round(random.uniform(1, 10), 2),
            "Delivery_Time_min": random.randint(15, 60),
            "Weather": random.choice(weather),
            "Traffic_Level": random.choice(traffic),
            "Food_Type": random.choice(food_type),
            "Delivery_Fee": round(random.uniform(20, 80), 2),
            "Customer_Rating": round(random.uniform(2.5, 5.0), 1)
        }

        dataset.append(record)

    return dataset

data = generate_dataset(100)

df = pd.DataFrame(data)
df.to_csv("food_delivery_dataset.csv", index=False)
print("✅ CSV file created!")


print("\n📊 Average Delivery Time by City:\n")
print(df.groupby("City")["Delivery_Time_min"].mean())

print("\n📊 Average Rating by Weather:\n")
print(df.groupby("Weather")["Customer_Rating"].mean())

print("\n📊 Delivery Fee Stats:\n")
print(df["Delivery_Fee"].describe())


print("\nMemory Usage:")
print(df.memory_usage(deep=True))

sample = data[0]

print("\nByte Representation:\n")
print("Integer:", struct.pack("i", sample["Order_ID"]))
print("Float:", struct.pack("f", sample["Delivery_Fee"]))
print("String:", sample["City"].encode("utf-8"))

print("\nObject Sizes:")
print("List:", sys.getsizeof(data))
print("Dict:", sys.getsizeof(sample))
```
